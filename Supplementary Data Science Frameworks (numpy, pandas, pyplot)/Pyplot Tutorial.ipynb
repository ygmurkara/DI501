<h3 style="margin-bottom:0">3) Plotting</h3>
<br>
<p style="margin-top:0; text-align: justify">Here, we will show you how to visualize using pyplot and seaborn. Here are a lot of options you can do with your graphs. We will present you some basics. </p>

First, if you want to plot some variable against another, you can simply use plt.plot. Here, we plotted our array "a" against array "b". You may omit plt.show() here in Jupyter Notebook, you can still get the graph. However, for example, if you work on Visual Studio Code, you cannot see your graph unless you explicitly write plt.show(). 

```python
plt.plot(a, b)
plt.show()
```

Now let's have a look at how we can play with this figure. We can add label names, play with axis sizes, change line type and color, its tranparency, width of the line etc. by: 

```python
plt.plot(a, b, 'mo:', alpha=0.3, linewidth=3)
plt.axis([0, 90, -1, 1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

"mo:" seems strange here. Here, m means color "magenta", o means "with marker" and : means "dashed line". To see other alternatives, you can check <a href="https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html">this website.</a> Alpha is used to determine transparency. With linewidth, we defined our line width. In plt.axis, we first defined our x axis limits then we passed our y axis limits. Finally, we gave names with xlabel and ylabel functions.

You can plot several lines with different format styles in one function call using arrays.

```python
plt.plot(a, b, 'ro', a*2, b, 'm:', a*2, b, 'g^')
plt.show()
```

You can craete scatter plots with 'o', but we also have a plt.scatter to create them easier.

```python
plt.scatter(x,y)
plt.show()
```

<h3 style="margin-bottom:0">4) Pyplot with Dataset</h3>
<br>
<p style="margin-top:0; text-align: justify">Here, we will show you how you can use pyplot on datasets. We will pick a dataframe and have a look at different ways to query data. I will quickly create a dataset which contains 25 people with their age, gender and a test score.</p>

```python
data = np.array([[18, 24, 26, 24, 20, 29, 32, 24, 21, 20, 28, 27, 20, 19, 31, 30, 25, 26, 24, 21, 29, 28, 27, 21, 21],['M','F','F','M','F','M','F','F','M','F','M','F','F','M','F','M','F','F','M','F','M','F','F','M','F'],[78, 89, 52, 89, 85, 48, 15, 27, 39, 42, 58, 79, 88, 89, 80, 78, 79, 89, 98, 65, 62, 31, 68, 79, 88]])

df = pd.DataFrame(data = data, index = ['age','gender','score']).transpose()
df.head()
```

```python
df.dtypes
```

As you can see, when we create pandas dataframe this way, we may encounter with problems such as wrong data types. We need to correct them before doing any analysis by: 

```python
df['age'] = df['age'].astype('int32')
df['score'] = df['score'].astype('int32')
df['gender'] = df['gender'].astype('string')
```

```python
df.dtypes
```

So, we can look at the histogram of our ages.

```python
plt.hist(df['age'])
plt.show()
```

If you do not like the number of bins, you can change it by (default is 10):

```python
plt.hist(df['age'], bins=6)
plt.show()
```

We can draw a pie chart as well. To draw a pie chart, we can use plt.pie.

```python
plt.pie(df['score'], radius=2, labels=df['score'])
plt.show()
```

Here, we used radius as 2 (default is 1) to visualize better and also gave label names with labels=df['score'] parameter.

We can also look at boxplot of age and score data as follows:

```python
plt.boxplot(df['age'])
plt.show()
```

Here, in bloxplot we can learn several useful properties: 
<ul><li> Box in the middle shows Q3 (above line) and Q1 (below line). So, their difference is interquartile range.</li>
    <li> Orange line in the middle is median.</li>
    <li> Lines at the top and at the bottom indicate max and min values, respectively.</li>

We can also draw scatter plot of age and score. However, here we can add something more and color individiual dots based on gender as well. This is much easier to do in seaborn, however we will show you how to do it in pyplot as well.

```python
colors = {'M':'tab:blue', 'F':'tab:orange'}
plt.scatter(df['age'],df['score'], c=df['gender'].map(colors))
```
